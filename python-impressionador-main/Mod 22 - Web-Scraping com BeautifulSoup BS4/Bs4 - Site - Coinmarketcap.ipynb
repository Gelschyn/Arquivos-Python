```python
from bs4 import BeautifulSoup
import requests

link = 'https://coinmarketcap.com/'
requisicao = requests.get(link)
site = BeautifulSoup(requisicao.text, 'html.parser')
```

```python
import re

tbody = site.find('tbody')
linhas = tbody.find_all('tr')
for linha in linhas:
    try:
        print(linha.find(class_='kKpPOn').text)
        print(linha.find(string=re.compile('\$')))
        print()
    except AttributeError:
        print('Todas as informações carregadas exibidas.')
        break
```

```python
import re

tbody = site.find('tbody')
linhas = tbody.find_all('tr')
for linha in linhas:
    texto_linha = linha.get_text(separator=';')
    lista_textos = texto_linha.split(';')
    try:
        nome = lista_textos[1]
        preco = lista_textos[4]
        print(nome, preco, '\n')
    except:
        nome = lista_textos[0]
        preco = '$' + lista_textos[3]
        print(nome, preco, '\n')
```
