#### 2. Faça um Programa que crie uma lista com as médias de cada aluno, imprima as médias de cada aluno e a quantidade de alunos com média maior que 7.

#### 4.Em quais meses a média de temperatura foi maior do que a média nacional?

````python
meses = [
    'Janeiro',
    'Fevereiro',
    'Março',
    'Abril',
    'Maio',
    'Junho',
    'Julho',
    'Agosto',
    'Setembro',
    'Outubro',
    'Novembro',
    'Dezembro'
]

temperaturas = [30, 29, 28, 28, 25, 26, 20, 21, 19, 25, 27, 32]

media_temperatura = sum(temperaturas)/len(temperaturas)

meses_maior = []

for i, mes in enumerate(meses):
    if temperaturas[i] > media_temperatura:
        meses_maior.append(mes)

if len(meses_maior) == 1:
    print(f'O mês de {meses_maior[0]} teve a média de temperatura maior que a média anual.')
else:
    nomes_meses = '; '.join(meses_maior)
    print(f'Os meses que tiveram a média de temperatura maior que a média anual foram: {nomes_meses}.')
````

7. Resolvendo com Listas de Listas:

````python
# Define a função que calcula o salário
def calcular_salario(venda_bruta):
    salario = 200 + 0.09 * venda_bruta
    return salario

# Constantes: vendas e as faixas salariais
vendas = [1000, 2000, 3000, 4000, 5000, 6000, 5500, 4500, 3600]

faixas = [
    ['R$200 - R$299', 200, 299],
    ['R$300 - R$399', 300, 399],
    ['R$400 - R$499', 400, 499],
    ['R$500 - R$599', 500, 599],
    ['R$600 - R$699', 600, 699],
    ['R$700 - R$799', 700, 799],
    ['R$800 - R$899', 800, 899],
    ['R$900 - R$999', 900, 999],
    ['R$1000 ou mais', 1000, float('inf')]
]

# Inicializa os contadores
contadores = [0] * len(faixas)

# Itera sobre as vendas e incrementa os contadores das faixas salariais correspondentes
for venda in vendas:
    salario = calcular_salario(venda)
    for i in range(len(faixas)):
        if salario >= faixas[i][1] and salario <= faixas[i][2]:
            contadores[i] += 1
            break

# Imprime os resultados
print(f'{"Salários":^16} - {"Vendedores":^12}')
for i in range(len(contadores)):
    print(f'{faixas[i][0]:^16} - {contadores[i]:^12}')
````

7. Resolvendo com Dicionário:

````python
# Define a função que calcula o salário
def calcular_salario(venda_bruta):
    salario = 200 + 0.09 * venda_bruta
    return salario

# Constante vendas:
vendas = [1000, 2000, 3000, 4000, 5000, 6000, 5500, 4500, 3600]

# Define as faixas salariais
faixas = {
    f"R${valor} - R${valor+99}": pos for pos, valor in enumerate(range(200, 1000, 100))
}
faixas["R$1000 ou mais"] = 8

# Inicializa os contadores
contadores = [0] * len(faixas)

# Itera sobre as vendas e incrementa os contadores das faixas salariais correspondentes
for venda in vendas:
    salario = calcular_salario(venda)
    for faixa, posicao in faixas.items():
        # Extrai os valores mínimo e máximo da faixa salarial
        minimo, maximo = map(int, faixa.replace('R$', '').strip().split("-"))
        # Se o salário está dentro da faixa, incrementa o contador e interrompe a iteração
        if minimo <= salario <= maximo:
            contadores[posicao] += 1
            break

# Imprime os resultados
print(f"{'Salários':^16} - {'Vendedores':^12}")
for i, contador in enumerate(contadores):
    faixa = list(faixas.keys())[i]
    print(f"{faixa:^16} - {contador:^12}")
````

7. Resolvendo com Tuplas

````python
# Define a função que calcula o salário
def calcular_salario(venda_bruta):
    salario = 200 + 0.09 * venda_bruta
    return salario

# Constante vendas:
vendas = [1000, 2000, 3000, 4000, 5000, 6000, 5500, 4500, 3600]

# Define as faixas salariais
faixas = [
    (range(200, 300), 'R$200 - R$299'),
    (range(300, 400), 'R$300 - R$399'),
    (range(400, 500), 'R$400 - R$499'),
    (range(500, 600), 'R$500 - R$599'),
    (range(600, 700), 'R$600 - R$699'),
    (range(700, 800), 'R$700 - R$799'),
    (range(800, 900), 'R$800 - R$899'),
    (range(900, 1000), 'R$900 - R$999'),
    (range(1000, 999999), 'R$1000 ou mais')
]

# Inicializa os contadores
contadores = [0] * len(faixas)

# Itera sobre as vendas e incrementa os contadores das faixas salariais correspondentes
for venda in vendas:
    salario = calcular_salario(venda)
    for faixa, faixa_str in faixas:
        if salario >= faixa.start and salario < faixa.stop:
            contadores[faixas.index((faixa, faixa_str))] += 1
            break

# Imprime os resultados
print(f"{'Salários':^16} - {'Vendedores':^12}")
for i, contador in enumerate(contadores):
    print(f'{faixas[i][1]:^16} - {contadores[i]:^12}')
````
