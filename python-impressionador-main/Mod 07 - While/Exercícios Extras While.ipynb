#### 1. Faça um programa que peça uma nota, entre zero e dez. Mostre uma mensagem caso o valor seja inválido e continue pedindo até que o usuário informe um valor válido.

````python
while True:
    try:
        nota = input('Digite uma nota entre 0 e 10 (ou "q" para sair): ')
        if nota.lower() == 'q':
            print('Encerrando programa...')
            break
        nota = float(nota)
        if nota >=0 and nota <=10:
            print(f'A nota informada foi: {nota}.')
            break
        else:
            raise ValueError
    except ValueError:
        print('Entrada inválida. Insira novamente uma nota entre 0 e 10.')
````

#### 2. Faça um programa que leia um nome de usuário e a sua senha e não aceite a senha igual ao nome do usuário, mostrando uma mensagem de erro e voltando a pedir as informações.

````python
while True:
    nome_de_usuario = input('Insira um nome de usuário(Min. 3 caracteres): ').replace(" ", "").strip()

    if not nome_de_usuario or len(nome_de_usuario) < 3:
        print('Nome de usuário inválido. Nome de usuário precisa ter no mínimo 3 caracteres.).')
        continue

    print(f'Nome de usuário cadastrado: {nome_de_usuario}')
    break

while True:
    senha = input('Insira uma senha (Ela não pode ser igual ao nome de usuário - Min. 4 caracteres): ').replace(" ", "").strip()

    if not senha or len(senha) < 4:
        print('Senha inválida. A senha precisa ter no mínimo 4 caracteres.')
        continue

    if senha == nome_de_usuario:
        print('Senha inválida. A senha não pode ser igual ao nome de usuário.')
        continue

    print('Senha válida. Cadastro concluído.')
    break
````

#### 3. Faça um programa que leia e valide as seguintes informações (e para cada uma delas, continue pedindo a informação até o usuário inserir corretamente):
##### Nome: maior que 3 caracteres;
##### Idade: entre 0 e 150;
##### Salário: maior que zero;
##### Sexo: 'f' ou 'm';
##### Estado Civil: 's', 'c', 'v', 'd';

````python
#Função para tratar o nome inserido pelo usuário e verificar:
def formatar_verificar_nome(nome):
    nome_formatado = nome.replace("  ", " ").title()
    nome_verificado = nome.replace(" ", "")
    if nome_verificado.isalpha() and len(nome_verificado) > 3:
        return nome_formatado, True
    else:
        return None, False

#Função para verificar a idade
def verificar_idade(idade):
    if idade.isnumeric() and 0 < int(idade) <150:
        return idade, True
    else:
        return None, False
    
# Função para verificar o salário
def verificar_salario(salario):
    try:
        salario = float(salario)
        if salario > 0:
            return salario, True
        else:
            return None, False
    except:
        return None, False

#Função para verificar o gênero:
def verificar_genero(letra_genero):
    generos = {
    "H": "Homem",
    "M": "Mulher",
    "O": "Outros"
}
    if letra_genero in generos:
        return generos.get(letra_genero), True
    else:
        return None, False

#Função para verificar o Estado Civil:
def verificar_estado_civil(letra_estado_civil):
    estados_civis = {
    "C": "Casado",
    "S": "Solteiro",
    "D": "Divorciado",
    "V": "Viúvo",
    "O": "Outros"
}
    if letra_estado_civil in estados_civis:
        return estados_civis.get(letra_estado_civil), True
    else:
        return None, False

#Laços de repetição para coletar e informar os dados do usuário corretamente:
while True:
    nome = input(f'Insira o seu nome(Mínimo 3 caracteres): ')
    nome, validar_nome = formatar_verificar_nome(nome)
    if validar_nome:
        print(f'Nome cadastrado com sucesso: {nome}.')
        break
    else:
        print('Nome inválido. Por favor digite o nome novamente.')
        continue

while True:
    idade = input(f'Insira a sua idade: ')
    idade, validar_idade = verificar_idade(idade)
    if validar_idade:
        print(f'Idade cadastrada com sucesso: {idade}.')
        break
    else:
        print('Idade inválida. Por favor digite a idade novamente.')
        continue

while True:
    salario = input(f'Insira seu salário: ')
    salario, validar_salario = verificar_salario(salario)
    if validar_salario:
        print(f'Salário cadastrada com sucesso: R$ {salario:,.2f}.')
        break
    else:
        print('Salário inválido. Por favor digite seu salário novamente.')
        continue

while True:
    letra_genero = input('Informe o seu gênero(H para Homens, M para Mulheres ou O para Outros): ').upper()
    genero, validar_genero = verificar_genero(letra_genero)
    if validar_genero:
        print(f'Gênero cadastrado com sucesso: {letra_genero} - {genero}.')
        break
    else:
        print('Genero não consta no banco de dados. Por favor informe seu gênero novamente(H, M ou O).')
        continue

while True:
    letra_estado_civil = input('Insira a letra correspondente ao seu estado civil(C - Casado, S - Solteiro, D - Divorciado, V - Viúvo ou O - Outros): ').upper()
    estado_civil, validar_estado_civil = verificar_estado_civil(letra_estado_civil)
    if validar_estado_civil:
        print(f'Estado civil cadastrado: {letra_estado_civil} - {estado_civil}')
        break
    else:
        print('Opção inválida. Por favor, informe novamente seu estado civil(C, S, D, V, O).')
        continue
````

#### 9. Numa eleição existem três candidatos. Faça um programa que peça o número total de eleitores. Peça para cada eleitor votar e ao final mostrar o número de votos de cada candidato.

````python
def obter_quantidade_eleitores():
    while True:
        qtd_eleitores = input('Digite a quantidade de eleitores que irão votar: ')
        if qtd_eleitores.isdigit() and int(qtd_eleitores) > 0:
            qtd_eleitores = int(qtd_eleitores)
            print(f'Serão contabilizados: {qtd_eleitores} eleitores.')
            return qtd_eleitores
        print('Entrada inválida. Insira apenas números inteiros positivos.')

def obter_voto_valido():
    while True:
        numero_voto = input(f'Digite para quem irá seu voto (Candidatos: 1, 2 ou 3): ')
        if numero_voto.isdigit() and 1 <= int(numero_voto) <= 3:
            return int(numero_voto)
        print('Número de candidato inválido. Por favor escolha entre os candidatos 1, 2 ou 3.')

candidatos = {
    f"Candidato {numero}": 0 for numero in range(1, 4)
}

qtd_eleitores = obter_quantidade_eleitores()

for _ in range(qtd_eleitores):
    voto = obter_voto_valido()
    for candidato, votos in candidatos.items():
        if int(voto) == int(candidato.split()[1]): # verifica se o número do candidato corresponde ao voto
            candidatos[candidato] += 1

print(f'{"Candidatos":^14} - {"Votos":^8}')
for candidato, votos in candidatos.items():
    print(f'{candidato:^14} - {votos:^8}')
````

#### 10. Faça um programa que calcule o valor total investido por um colecionador em sua coleção de CDs e o valor médio gasto em cada um deles. O usuário deverá informar a quantidade de CDs e o valor para cada um.

````python
def obter_quantidade_cds():
    while True:
        qtd_cds = input('Digite a quantidade de CDs da coleção: ')
        if qtd_cds.isdigit() and int(qtd_cds) > 0:
            qtd_cds = int(qtd_cds)
            print(f'A coleção possui {qtd_cds} CDs.')
            return qtd_cds
        print('Entrada inválida. Insira apenas números inteiros positivos.')

def obter_valores_cds(num_cds):
    valores = []
    for i in range(num_cds):
        while True:
            valor = input(f'Digite o valor do {i+1}° CD: ')
            try:
                valor = float(valor)
                if valor > 0:
                    valores.append(valor)
                    break
            except ValueError:
                print('Valor inválido. Insira apenas números positivos.')
    return valores


qtd_cds = obter_quantidade_cds()
valores_cds = obter_valores_cds(qtd_cds)

valor_total = sum(valores_cds)
valor_medio = valor_total / qtd_cds

print(f'O valor total investido na coleção foi de R${valor_total:.2f}.')
print(f'O valor médio gasto em cada CD foi de R${valor_medio:.2f}.')
````

#### 13. O cardápio de uma lanchonete é o seguinte:
<pre>
Especificação   Código  Preço
Cachorro Quente 100     R$ 1,20
Bauru Simples   101     R$ 1,30
Bauru com ovo   102     R$ 1,50
Hambúrguer      103     R$ 1,20
Cheeseburguer   104     R$ 1,30
Refrigerante    105     R$ 1,00
Faça um programa que leia o código dos itens pedidos e as quantidades desejadas. 
Calcule e mostre o valor a ser pago por item (preço * quantidade) e o total geral do pedido. 
Considere que o cliente deve informar quando o pedido deve ser encerrado.
</pre>

````python
cardapio = {
    100: ('Cachorro Quente', 1.20),
    101: ('Bauru Simples', 1.30),
    102: ('Bauru com Ovo', 1.50),
    103: ('Hambúrguer', 1.20),
    104: ('Cheeseburguer', 1.30),
    105: ('Refrigerante', 1.00),
}

#Mostrando o cardápio para o usuário selecionar suas compras:
print('Cardápio Lanchonete:\n')
print(f'{"Código":^8}   {"Item":^18}   {"Preço":^8}')
for codigo, (item, preco) in cardapio.items():
    print(f'{codigo:^8} - {item:^18} - {f"R${preco:,.2f}":^8}')
print()
#Dicionário para registrar as compras:
dicionario_compras = {}

#Verifica e valida as informações inseridas pelo usuário e registra elas no dicionario_compras:
continuar = True
while continuar:
        while True:
            pedido = input('Digite o código do item que você gostaria de comprar: ')
            if pedido.isnumeric() and int(pedido) in cardapio:
                pedido = int(pedido)
                while True:
                    quantidade = input(f'Digite quantos "{cardapio[pedido][0]}" você quer comprar ou "V" para voltar: ').upper()
                    if quantidade.isnumeric():
                        quantidade = int(quantidade)
                        # Verifica se o item já existe no dicionário de compras
                        if pedido in dicionario_compras:
                            # Se sim, soma a nova quantidade com a quantidade já existente
                            dicionario_compras[pedido] = (
                                cardapio[pedido][0],
                                cardapio[pedido][1],
                                dicionario_compras[pedido][2] + quantidade,
                                (dicionario_compras[pedido][2] + quantidade) * cardapio[pedido][1]
                            )
                        else:
                            # Se não, adiciona o novo item com a quantidade informada pelo usuário
                            dicionario_compras[pedido] = (
                                cardapio[pedido][0],
                                cardapio[pedido][1],
                                quantidade,
                                quantidade * cardapio[pedido][1]
                            )
                        break
                    elif quantidade == 'V':
                        # Usuário escolheu voltar para o laço de código de pedido
                        break
                    else:
                        print('Quantidade inválida, por favor digite um valor positivo ou 0 caso tenha inserido o item por acidente.')
                        continue
                break    
            else:
                print('Código do item informado não consta no cardápio. Por favor insira o código como é exibido no cardápio.')
                continue
        while True:    
            continuar_compra = input('Gostaria de continuar comprando? (Digite "S" para Sim ou "N" para Não): ').upper().strip()
            if continuar_compra == "S":
                continuar = True
                break
            elif continuar_compra == "N":
                continuar = False
                break
            else:
                print('Por favor, responda apenas com "S" para sim e "N" para não.')
                continue
        
            

#Exibindo as compras e o valor final para o usuário:
print('\nNota fiscal:')
print('-'*74)
print(f'{"Código":^8}   {"Item":^18}   {"Preço Un.":^10}   {"Quantidade":^12}   {"Preço Total":^14}')
for codigo, (item, preco, qtd, preco_qtd) in dicionario_compras.items():
    print(f'{codigo:^8} - {item:^18} - {f"R${preco:,.2f}":^10} - {f"{qtd}un.":^12} - {f"R${preco_qtd:,.2f}":^14}')
print()

valor_final = sum(compra[3] for compra in dicionario_compras.values())
print(f'Valor Total Final: R${valor_final:,.2f}')
````

Para isto, faz-se necessário o desenvolvimento de um programa, que será utilizado pelas telefonistas, para a computação dos votos. Sua equipe foi contratada para desenvolver este programa, utilizando a linguagem de programação Python. Para computar cada voto, a telefonista digitará um número, entre 1 e 23, correspondente ao número da camisa do jogador. Um número de jogador igual zero, indica que a votação foi encerrada. Se um número inválido for digitado, o programa deve ignorá-lo, mostrando uma breve mensagem de aviso, e voltando a pedir outro número. Após o final da votação, o programa deverá exibir:
<pre>a. O total de votos computados;
b. Os números e respectivos votos de todos os jogadores que receberam votos;
c. O percentual de votos de cada um destes jogadores;
d. O número do jogador escolhido como o melhor jogador da partida, juntamente com o número de votos e o percentual de votos dados a ele.
        Observe que os votos inválidos e o zero final não devem ser computados como votos. O resultado aparece ordenado pelo número do jogador. O programa deve fazer uso de arrays. O programa deverá executar o cálculo do percentual de cada jogador através de uma função. Esta função receberá dois parâmetros: o número de votos de um jogador e o total de votos. A função calculará o percentual e retornará o valor calculado. Abaixo segue uma tela de exemplo. O disposição das informações deve ser o mais próxima possível ao exemplo. Os dados são fictícios e podem mudar a cada execução do programa. Ao final, o programa deve ainda gravar os dados referentes ao resultado da votação em um arquivo texto no disco, obedecendo a mesma disposição apresentada na tela.</pre>

````python
votos = [0] * 23
melhor = [(0, 0)]
contagem = 0

print('Enquete: Quem foi o melhor jogador?')

while True:
    jogador = int(input('Número do jogador (0 = fim): '))
    if jogador == 0:
        break
    elif 1 > jogador > 23:
        print('Informe um valor entre 1 e 23 ou 0 para sair!')
        continue
    votos[jogador - 1] += 1

total = sum(votos)
if total == 0:
    print ('Não tivemos votos computados.')
else:
    print('\nResultado da votação: \n')
    print(f'Foram computados {total} votos.\n')
    print('Jogador\t\tVotos\t\t%')
    for i, voto in enumerate(votos):
        if voto > 0:
            print(f'{i+1}\t\t{voto}\t\t{voto/total:.2%}')
            if voto > contagem:
                contagem = voto
                melhor[0] = ((int(i+1), contagem))
            elif voto == contagem:
                contagem = voto
                melhor.append((int(i+1), contagem))
    
    if len(melhor) > 1:
        print('\nTivemos empate entre os seguintes jogadores:')
        for numero, contagem in melhor:
            print(f'Jogador de número {numero}, com {contagem} votos, correspondendo a {(contagem/total):.2%} do total de votos.')
    else:
        for numero, contagem in melhor:
            print(f'\nO melhor jogador foi o de número {numero}, com {contagem} votos, correspondendo a {(contagem/total):.2%} do total de votos.')
````

````python
# Inicializa um array com 23 posições, cada uma representando um jogador
votos = [0] * 23

# Inicializa uma lista para armazenar os jogadores com maior contagem de votos
melhores_jogadores = []

while True:
    jogador = int(input('Número do jogador (0 = fim): '))
    if jogador == 0:
        break
    elif jogador < 1 or jogador > 23:
        print('Informe um valor entre 1 e 23 ou 0 para sair!')
        continue
    votos[jogador - 1] += 1

# Calcula o total de votos
total_votos = sum(votos)

if total_votos == 0:
    print('Não houve votos computados.')
else:
    print('\nResultado da votação:\n')
    print(f'Foram computados {total_votos} votos.\n')
    print('Jogador\t\tVotos\t\t%')

    # Percorre o array de votos e imprime apenas os jogadores que receberam votos
    for i, voto in enumerate(votos):
        if voto > 0:
            percentual = voto / total_votos * 100
            print(f'{i+1}\t\t{voto}\t\t{percentual:.1f}%')
            
            # Se o jogador atual tiver mais votos que o(s) melhor(es) jogador(es) anterior(es),
            # atualiza a lista de melhores jogadores
            if not melhores_jogadores or voto > melhores_jogadores[0][1]:
                melhores_jogadores = [(i+1, voto)]
            elif voto == melhores_jogadores[0][1]:
                melhores_jogadores.append((i+1, voto))

    # Imprime o(s) melhor(es) jogador(es)
    if len(melhores_jogadores) == 1:
        numero, contagem = melhores_jogadores[0]
        percentual = contagem / total_votos * 100
        print(f'\nO melhor jogador foi o de número {numero}, com {contagem} votos, correspondendo a {percentual:.1f}% do total de votos.')
    else:
        print('\nTivemos empate entre os seguintes jogadores:')
        for numero, contagem in melhores_jogadores:
            percentual = contagem / total_votos * 100
            print(f'Jogador de número {numero}, com {contagem} votos, correspondendo a {percentual:.1f}% do total de votos.')
````

````python
from collections import defaultdict

VOTOS_TOTAIS = "Foram computados {} votos."
RESULTADO_VOTACAO = "Resultado da votação:\n"
HEADER_TABELA = "Jogador\t\tVotos\t\t%"
MELHOR_JOGADOR = "O melhor jogador foi o de número {}, com {} votos, correspondendo a {:.1f}% do total de votos."
EMPATE = "Tivemos empate entre os seguintes jogadores:"

votos = defaultdict(int)

while True:
    jogador = int(input("Número do jogador (0 = fim): "))
    if jogador == 0:
        break
    elif jogador < 1 or jogador > 23:
        print("Informe um valor entre 1 e 23 ou 0 para sair!")
        continue
    votos[jogador] += 1

total_votos = sum(votos.values())

if total_votos == 0:
    print("Não houve votos computados.")
else:
    print(VOTOS_TOTAIS.format(total_votos))
    print(RESULTADO_VOTACAO)
    print(HEADER_TABELA)

    melhores_jogadores = []

    for jogador, votos_jogador in votos.items():
        if votos_jogador > 0:
            percentual = votos_jogador / total_votos * 100
            print(f"{jogador}\t\t{votos_jogador}\t\t{percentual:.1f}%")

            if not melhores_jogadores or votos_jogador > melhores_jogadores[0][1]:
                melhores_jogadores = [(jogador, votos_jogador)]
            elif votos_jogador == melhores_jogadores[0][1]:
                melhores_jogadores.append((jogador, votos_jogador))

    if len(melhores_jogadores) == 1:
        numero, contagem = melhores_jogadores[0]
        percentual = contagem / total_votos * 100
        print(MELHOR_JOGADOR.format(numero, contagem, percentual))
    else:
        print(EMPATE)
        for numero, contagem in melhores_jogadores:
            percentual = contagem / total_votos * 100
            print(f"Jogador de número {numero}, com {contagem} votos, correspondendo a {percentual:.1f}% do total de votos.")
````
