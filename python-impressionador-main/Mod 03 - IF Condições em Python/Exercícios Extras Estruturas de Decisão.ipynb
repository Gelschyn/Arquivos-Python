#### 12 . Faça um programa para o cálculo de uma folha de pagamento, sabendo que os descontos são do Imposto de Renda, que depende do salário bruto (conforme tabela abaixo) e que o FGTS corresponde a 11% do Salário Bruto, mas não é descontado (é a empresa que deposita). O Salário Líquido corresponde ao Salário Bruto menos os descontos. O programa deverá pedir ao usuário o valor da sua hora e a quantidade de horas trabalhadas no mês.
<pre>
Desconto do IR:<br>
Salário Bruto até 900 (inclusive) - isento<br>
Salário Bruto até 1500 (inclusive) - desconto de 5%<br>
Salário Bruto até 2500 (inclusive) - desconto de 10%<br>
Salário Bruto acima de 2500 - desconto de 20%<br>
Imprima na tela as informações, dispostas conforme o exemplo abaixo. No exemplo o valor da hora é 5 e a quantidade de hora é 220.

Salário Bruto: (5 * 220)        : R$ 1100,00<br>
(-) IR (5%)                     : R$   55,00<br>
(-) INSS ( 10%)                 : R$  110,00<br>
FGTS (11%)                      : R$  121,00<br>
Total de descontos              : R$  165,00<br>
Salário Liquido                 : R$  935,00<br>
</pre>
###### Obs: Não vamos nos preocupar tanto com a formatação dos números (nº de casas decimais, por exemplo, veremos isso no próximo módulo)


```python
#Constantes - Impostos em %:
inss = 10
fgts = 11

#Informando o valor das horas:
valor_hora = None
while not valor_hora:
    try:
        valor_hora = float(input('Insira o valor da sua hora de trabalho: R$'))
    except ValueError:
        print('Valor inválido. Insira apenas números.')

#Informando a quantidade de horas trabalhadas:
quantidade_horas = None
while not quantidade_horas:
    try:
        quantidade_horas = float(input('Insira quantas horas você trabalhou esse mês: '))
    except ValueError:
        print('Valor inválido. Insira apenas números.')

#Calculando o salário bruto:
salario_bruto = valor_hora * quantidade_horas

#Calculando o Imposto de Renda:
if salario_bruto > 2500:
    ir = 20
elif salario_bruto > 1500:
    ir = 10
elif salario_bruto > 900:
    ir = 5
else:
    ir = 0

#Calculando os descontos:
descontos = (salario_bruto * ir/100) + (salario_bruto * inss/100)

#Calculando o salário líquido:
salario_liquido = salario_bruto - descontos

#Exibindo as informações:
exibir_info = f'''
Salário Bruto: ( R${valor_hora:,.2f} * {quantidade_horas}h): R${salario_bruto:,.2f}
(-) IR ({ir}%): R${(salario_bruto * ir/100):,.2f}
(-) INSS ( 10%): R${(salario_bruto * inss/100):,.2f}
FGTS (11%): R${(salario_bruto * fgts/100):,.2f}
Total de descontos: R${descontos:,.2f}
Salário Liquido: R${salario_liquido:,.2f}
'''

#Printando pro usuário:
print(exibir_info.ljust(50))
```

Faça um Programa que peça um número correspondente a um determinado ano e em seguida informe se este ano é ou não bissexto.
<pre>
Dica para determinar se um ano é bissexto: 
- São bissextos todos os anos múltiplos de 400, p.ex.: 1600, 2000, 2400, 2800...
- São bissextos todos os múltiplos de 4, exceto se for múltiplo de 100 mas não de 400, 
p.ex.: 1996, 2000, 2004, 2008, 2012, 2016, 2020, 2024, 2028...
- Não são bissextos todos os demais anos.<br>
ex1: 2004 é múltiplo de 4, mas não é múltiplo de 100, então é bissexto.
ex2: 2000 é múltiplo de 4, mas é múltiplo de 100, só que também é multiplo de 400, então é bissexto (porque todo ano múltiplo de 400 é bissexto, independente do resto).
ex3: 1900 é múltiplo de 4, é múltiplo de 100, mas não é múltiplo de 400, então não é bissexto

</pre>

Dica: lembre que: numero % 4 é o resto da divisão do número por 4, ex: 10 % 3 = 1 (já que 10/3 = 3 e resta 1)

```python
ano = input('Digite o ano: ')

# Verificando se a entrada do usuário é um número inteiro e positivo:
while not ano.isdigit() or int(ano) <= 0:
    print('Entrada inválida. O ano deve ser um número inteiro e positivo.')
    ano = input('Digite o ano novamente: ')

ano = int(ano)

if ano % 400 == 0 or (ano % 4 == 0 and ano % 100 != 0):
    print(f'{ano} é um ano bissexto.')
else:
    print(f'{ano} não é um ano bissexto.')
```

#### 18. Faça um Programa para um caixa eletrônico. O programa deverá perguntar ao usuário a valor do saque e depois informar quantas notas de cada valor serão fornecidas. As notas disponíveis serão as de 1, 5, 10, 50 e 100 reais. O valor mínimo é de 10 reais e o máximo de 600 reais. O programa não deve se preocupar com a quantidade de notas existentes na máquina.
<pre>
Exemplo 1: Para sacar a quantia de 256 reais, o programa fornece duas notas de 100, uma nota de 50, uma nota de 5 e uma nota de 1;
Exemplo 2: Para sacar a quantia de 399 reais, o programa fornece três notas de 100, uma nota de 50, quatro notas de 10, uma nota de 5 e quatro notas de 1.
</pre>

Dica1: numero // 10 vai te dar como resposta a parte inteira da divisão do número por 10.<br>
Dica2: numero % 10 vai te dar o resto da divisão do número por 10.

```python
notas = {100: 'R$100', 50: 'R$50', 10: 'R$10', 5: 'R$5', 1: 'R$1'}

def calcular_notas(valor, notas):
    saque = {}
    for nota in notas:
        qtd_notas = valor // nota
        valor = valor % nota
        if qtd_notas > 0:
            saque[nota] = qtd_notas
    return saque

while True:
    try:
        valor_saque = float(input('Insira o valor do saque entre R$10 e R$600: '))
        if valor_saque < 10 or valor_saque > 600:
            raise ValueError
        break
    except ValueError:
        print('Valor de saque inválido. Insira apenas valores entre R$10 e R$600.')

notas_saque = calcular_notas(valor_saque, notas)
print('Serão fornecidas:')
for nota, qtd in notas_saque.items():
    print(f'{int(qtd)} nota(s) de {notas[nota]}.')
```

```python
notas = [100, 50, 10, 5, 1]

def calcular_notas(valor):
    saque = []
    for nota in notas:
        qtd_notas = valor // nota
        valor = valor % nota
        saque.append(qtd_notas)
    return saque

while True:
    try:
        valor_saque = float(input('Insira o valor do saque entre R$10 e R$600: '))
        if valor_saque < 10 or valor_saque > 600:
            raise ValueError
        break
    except ValueError:
        print('Valor de saque inválido. Insira apenas valores entre R$10 e R$600.')

notas_saque = calcular_notas(valor_saque)
print('Serão fornecidas:')
for i, qtd in enumerate(notas_saque):
    if qtd > 0:
        print(f'{int(qtd)} nota(s) de R${notas[i]:.2f}')
```

#### 19. Faça um programa que faça 5 perguntas para uma pessoa sobre um crime. As perguntas são:
<pre>
"Telefonou para a vítima?"
"Esteve no local do crime?"
"Mora perto da vítima?"
"Devia para a vítima?"
"Já trabalhou com a vítima?" 
O programa deve no final emitir uma classificação sobre a participação da pessoa no crime. 
Se a pessoa responder positivamente a 2 questões ela deve ser classificada como "Suspeita",
entre 3 e 4 como "Cúmplice" e 5 como "Assassino". 
Caso contrário, ele será classificado como "Inocente".
</pre>

```python
print('Responda às perguntas abaixo sobre o crime apenas com "Sim" ou "Não".')

perguntas = ['Telefonou para a vítima?', 'Esteve no local do crime?', 'Mora perto da vítima?', 'Devia para a vítima?', 'Já trabalhou com a vítima?']

respostas = []
for pergunta in perguntas:
    resposta = input(pergunta)
    while resposta.lower() not in ['sim', 'não']:
        print('Por favor, responda apenas com "Sim" ou "Não".')
        resposta = input(pergunta)
    respostas.append(resposta.lower())

respostas_positivas = respostas.count('sim')

if respostas_positivas == 5:
    print('Você foi classificado como "Assassino".')
elif respostas_positivas >= 3:
    print('Você foi classificado como "Cúmplice".')
elif respostas_positivas == 2:
    print('Você foi classificado como "Suspeito".')
else:
    print('Você foi classificado como "Inocente".')
```

```python
preco_frutas = {
    'morango': {
        'preco_kg': 2.5,
        'preco_kg_acima': 2.2,
    },
    'maca': {
        'preco_kg': 1.8,
        'preco_kg_acima': 1.5,
    }
}

def calcular_preco_frutas(tipo_fruta, kg):
    if kg <= 5:
        preco_kg = preco_frutas[tipo_fruta]['preco_kg']
    else:
        preco_kg = preco_frutas[tipo_fruta]['preco_kg_acima']
    valor_fruta = kg * preco_kg
    return valor_fruta

def calcular_desconto(total_compra):
    if total_compra > 25 or total_kg > 8:
        desconto = total_compra * 0.1
    else:
        desconto = 0
    return desconto

while True:
    try:
        kg_morango = float(input('Quantos Kg de morangos? '))
        kg_maca = float(input('Quantos Kg de maçãs? '))
        if kg_morango < 0 or kg_maca < 0:
            raise ValueError
        break
    except ValueError:
        print('Insira novamente o peso dos morangos e das maças, somente valores positivos.')

total_kg = kg_morango + kg_maca
valor_frutas = calcular_preco_frutas('morango', kg_morango) + calcular_preco_frutas('maca', kg_maca)
valor_compra = valor_frutas - calcular_desconto(valor_frutas)

print(f'Peso total das frutas: {total_kg:.2f}Kg')

if calcular_desconto(valor_frutas) == 0:
    print(f'Essa compra não teve desconto e o valor total a ser pago será de: R${valor_frutas:,.2f}')
else:
    print(f'O valor da compra foi de: R${valor_frutas:,.2f}. E após aplicar o desconto ficou de: R$ {valor_compra:,.2f}')
```

Se compra for feita no cartão Tabajara o cliente receberá ainda um desconto de 5% sobre o total da compra. Escreva um programa que peça o tipo e a quantidade de carne comprada pelo usuário e gere um cupom fiscal, contendo as informações da compra: tipo e quantidade de carne, preço total, tipo de pagamento, valor do desconto e valor a pagar.

##### 3. Misturar latas e galões, de forma que o desperdício de tinta seja menor. Acrescente 10% de folga e sempre arredonde os valores para cima, isto é, considere latas cheias.


O custo da lata é 80/18 = 4,44 R$/L

O custo do galão é 25/3,6 = 6,94 R$/L

A lata é mais econômica, então todas as latas inteiras que pudermos usar devemos comprar em latas. Se ficar faltando alguma coisa para completar devemos avaliar se é melhor comprar latas ou galões. Exemplo:

Se queremos comprar 90 litros. 5 latas dão exatamente 90 litros. Então devemos comprar tudo em latas.

Se queremos comprar 95 litros. 5 latas dão exatamente 90 litros. Então devemos comprar pelo menos 5 latas e avaliar o que falta, se estes últimos 5 litros valem mais apenas em latas ou galões.

Para os 5 litros faltantes precisamos de 2 galões que custam 50 reais no total. Ou de uma lata que custa 80 reais no total. Portanto, neste caso vale mais a pena usar 2 galões.

Se queremos comprar 107 litros. 5 latas dão exatamente 90 litros. Então devemos comprar pelo menos 5 latas e avaliar o que falta, se estes últimos 5 litros valem mais apenas em latas ou galões.

Para os 17 litros faltantes precisamos de 5 galões que custam 125 reais no total. Ou de uma lata que custa 80 reais no total. Portanto, neste caso vale mais a pena usar uma lata.

3 galões custam 75 reais, 4 galões custam 100 reais. Então, se for possível completar com até 3 galões escolhe-se galões. Qualquer quantidade maior que 3 galões, usa-se latas.

Podemos ir ao exercício:

```python
import math

# Constantes:
cobertura_litro = 6
capacidade_lata = 18
preco_lata = 80
capacidade_galao = 3.6
preco_galao = 25

def calcular_litros_necessarios(area, cobertura_litro):
    litros_necessarios = area / cobertura_litro
    litros_necessarios *= 1.1 #10% de folga
    return litros_necessarios

while True:
    try:
        area = float(input('Informe o tamanho da área em metros quadrados que deverá ser pintada: '))
        if area <= 0:
            raise ValueError
        break
    except ValueError:
        print('Valor de área inválido. Insira apenas valores positivos.')

litros_necessarios = calcular_litros_necessarios(area, cobertura_litro)

latas_necessarias = litros_necessarios // capacidade_lata
resto_litros = litros_necessarios % capacidade_lata

if resto_litros > 0 and resto_litros <= capacidade_galao * 3:
    galoes_necessarios = math.ceil(resto_litros / capacidade_galao)
else:
    latas_necessarias += math.ceil(resto_litros / capacidade_lata)
    galoes_necessarios = 0

custo_latas = preco_lata * latas_necessarias
custo_galoes = preco_galao * galoes_necessarios

if galoes_necessarios > 0:
    print(f'Serão necessários {int(latas_necessarias)} lata(s) de tinta e {int(galoes_necessarios)} galão(ões) de tinta.')
else:
    print(f'Será(ão) necessária(s) {int(latas_necessarias)} lata(s) de tinta.')

print(f'O custo total será de R${custo_latas + custo_galoes:,.2f}.')
```
