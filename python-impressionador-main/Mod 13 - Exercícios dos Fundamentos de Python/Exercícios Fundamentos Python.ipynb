### 1- Imposto a pagar no Lucro Presumido

- 5% sobre faturamento de ISS (mensal)
- 0,65% de PIS sobre faturamento, (mensal)
- 3% de COFINS sobre faturmaneto, (mensal)
- 4.8% de IR (trimestral)
- 10% de IR Adicional sobre o que ultrapassar 20mil do faturamento (trimestral)
- CSLL: 2,88% sobre faturamento (trimestral)

```python
faturamento = {
    'jan': 'R$ 95.141,98',
    'fev': 'R$ 95.425,16',
    'mar': 'R$ 89.716,31',
    'abr': 'R$ 78.459,99',
    'mai': 'R$ 71.087,28',
    'jun': 'R$ 83.911,06',
    'jul': 'R$ 56.467,26',
    'ago': 'R$ 88.513,58',
    'set': 'R$ 66.552,49',
    'out': 'R$ 80.164,07',
    'nov': 'R$ 66.964,33',
    'dez': 'R$ 71.525,25',
}

# você precisa inserir no sistema um dicionário no formato:

# resultado = {
#     'mes': (faturamento, imposto_mensal, imposto_trimestral),
# }
```

```python
#Função Calcular Impostos Mensais:
def calcular_imp_men(valor):
    iss = valor * 0.05
    pis = valor * 0.0065
    cofins = valor * 0.03
    imposto_mensal = iss + pis + cofins
    return imposto_mensal

#Função Calcilar Impostos Trimestrais:
def calcular_imp_tri(valor):
    ir_tri = valor * 0.048
    csll = valor * 0.0288
    ir_adc = ((valor - 20000) * 0.1) if valor > 20000 else 0
    imposto_tri = ir_tri + csll + ir_adc
    return imposto_tri

#Criando um novo dicionário com os valores numéricos:
dic_faturamento = {mes: float(valor.replace('R$', '').replace('.', '').replace(',', '.')) for mes, valor in faturamento.items()}

#Criando dicionário para armazenar o resultado:
dic_impostos = {}

#Calculando os valores e armazenando no dic_impostos
for mes, valor in dic_faturamento.items():
    dic_impostos[mes] = (valor, calcular_imp_men(valor), calcular_imp_tri(valor))

#Formatando o dicionário dic_impostos:
dic_impostos_formatado = {mes: (f"R$ {valor:,.2f}".replace(',', '-').replace('.', ',').replace('-', '.'), 
                        f"R$ {calcular_imp_men(valor):,.2f}".replace(',', '-').replace('.', ',').replace('-', '.'), 
                        f"R$ {calcular_imp_tri(valor):,.2f}".replace(',', '-').replace('.', ',').replace('-', '.')) 
                 for mes, valor in dic_faturamento.items()}

print(dic_impostos_formatado)
```

```python
#Dic de faturamento
faturamento = {
    'jan': 'R$ 95.141,98',
    'fev': 'R$ 95.425,16',
    'mar': 'R$ 89.716,31',
    'abr': 'R$ 78.459,99',
    'mai': 'R$ 71.087,28',
    'jun': 'R$ 83.911,06',
    'jul': 'R$ 56.467,26',
    'ago': 'R$ 88.513,58',
    'set': 'R$ 66.552,49',
    'out': 'R$ 80.164,07',
    'nov': 'R$ 66.964,33',
    'dez': 'R$ 71.525,25',
}

# Função para calcular impostos
def calcular_impostos(valor):
    aliquotas_impostos = {
    'iss': 0.05,
    'pis': 0.0065,
    'cofins': 0.03,
    'ir_tri': 0.048,
    'csll': 0.0288,
    'ir_adc': 0.1,
    'valor_ir_adc': 20000
}

    # Imposto mensal
    iss = valor * aliquotas_impostos['iss']
    pis = valor * aliquotas_impostos['pis']
    cofins = valor * aliquotas_impostos['cofins']
    impostos_mensal = iss + pis + cofins

    # Imposto trimestral
    ir_tri = valor * aliquotas_impostos['ir_tri']
    csll = valor * aliquotas_impostos['csll']
    ir_adc = ((valor - aliquotas_impostos['valor_ir_adc']) * aliquotas_impostos['ir_adc']) if valor > aliquotas_impostos['valor_ir_adc'] else 0
    impostos_trimestral = ir_tri + csll + ir_adc

    return impostos_mensal, impostos_trimestral

# Função para formatar valores de dinheiro
def formatar_valor(valor):
    return f"R$ {valor:,.2f}".replace('.', '-').replace(',', '.').replace('-', ',')

#Formatando o dic de faturamento:
dic_faturamento = {mes: float(valor.replace('R$', '').replace('.', '').replace(',', '.')) for mes, valor in faturamento.items()}

for mes, valor in dic_faturamento.items():
    imposto_mensal, imposto_trimestral = calcular_impostos(valor)
    dic_impostos[mes] = (formatar_valor(valor), formatar_valor(imposto_mensal), formatar_valor(imposto_trimestral))

# Impressão do resultado
print(dic_impostos)
```

### 3 - Extração dos links de Download de Vídeos do Vimeo

- Precisamos pegar os links em 1080p, 720p e 540p para importar os vídeos para uma nova plataforma

```python
#Para pegar o dicionario do vimeo, use:
#from dic import dicionario_vimeo
# você precisa chegar em:
# videos = [
#     {'uri': video_uri, 'nome': nome_video, 'duracao': duracao, 'link540p': link540p, 'link720p': link720p, 'link1080p': link1080p},
# ]
```

```python
# Importando o dicionário do vimeo:
from dic import dicionario_vimeo

# Criando a lista:
videos_download = []

# Percorrendo o dicionário e armazenando os dados necessários:
for dic_video in dicionario_vimeo['data']:
    video_info = {
        "uri": dic_video['uri'],
        "nome": dic_video['name'],
        "duração": dic_video['duration'],
        "link540p": '',
        "link720p": '',
        "link1080p": ''
    }
    for dic_download in dic_video['download']:
        if dic_download['rendition'] == '540p':
            video_info['link540p'] = dic_download['link']
        elif dic_download['rendition'] == '720p':
            video_info['link720p'] = dic_download['link']
        elif dic_download['rendition'] == '1080p':
            video_info['link1080p'] = dic_download['link']
    
    videos_download.append(video_info)

display(videos_download)
```
